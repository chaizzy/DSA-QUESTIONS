```python
# Ans 1:
def minimumDeleteSum(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0] * (n+1) for _ in range(m+1)]

    # Base cases
    for j in range(1, n+1):
        dp[0][j] = dp[0][j-1] + ord(s2[j-1])
    for i in range(1, m+1):
        dp[i][0] = dp[i-1][0] + ord(s1[i-1])

    # Fill the dp array
    for i in range(1, m+1):
        for j in range(1, n+1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1]
            else:
                dp[i][j] = min(dp[i-1][j] + ord(s1[i-1]), dp[i][j-1] + ord(s2[j-1]))

    return dp[m][n]
```

```python
# ANS 2:
def checkValidString(s):
    L = []
    
    for char in s:
        if char == '(' or char == '*':
            L.append(char)
        elif char == ')':
            if L and L[-1] == '(':
                L.pop()
            elif L and L[-1] == '*':
                L.pop()
            else:
                return False
    
    while L:
        if L[-1] == '(':
            return False
        L.pop()
    
    return True
```

```python
# ANS 3:
def minSteps(word1, word2):
    m, n = len(word1), len(word2)
    
    
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    
    for i in range(m + 1):
        dp[i][0] = i
    for j in range(n + 1):
        dp[0][j] = j
    
    
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if word1[i - 1] == word2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = min(dp[i - 1][j], dp[i][j - 1]) + 1
    
    return dp[m][n]
```

```python
# ANS 4:
def compress(chars):
    write = 0
    read = 0
    count = 1
    
    while read < len(chars):
        if read + 1 == len(chars) or chars[read] != chars[read + 1]:
            chars[write] = chars[read]
            write += 1
            if count > 1:
                for digit in str(count):
                    chars[write] = digit
                    write += 1
            count = 1
        else:
            count += 1
        read += 1
    
    return write
```

```python
# ANS 5:
from collections import Counter

def findAnagrams(s, p):
    freq_p = Counter(p)
    freq_window = Counter()
    result = []
    left = 0
    right = 0

    while right < len(s):
        freq_window[s[right]] += 1
        if right - left + 1 == len(p):
            if freq_window == freq_p:
                result.append(left)
            freq_window[s[left]] -= 1
            if freq_window[s[left]] == 0:
                del freq_window[s[left]]
            left += 1
        right += 1

    return result
```

```python
# ANS 7:
def decodeString(s):
    stack = []

    for c in s:
        if c.isdigit():
            stack.append(int(c))
        elif c.isalpha():
            stack.append(c)
        elif c == "[":
            stack.append("[")
        elif c == "]":
            substring = ""
            while stack[-1] != "[":
                substring = stack.pop() + substring
            stack.pop()  # Pop the opening bracket
            count = stack.pop()
            stack.append(substring * count)

    decoded_string = "".join(stack[::-1])
    return decoded_string
```
