```python
# ANS 1:
def construct2DArray(original, m, n):
    if len(original) != m * n:
        return []
    
    result = [[0] * n for _ in range(m)]
    
    for i in range(len(original)):
        row = i // n
        col = i % n
        result[row][col] = original[i]
    
    return result

original=[1,2,3,4]
m=2
n=2

output=construct2DArray(original,m,n)
print(output)
```

```python
# ANS 2
def find_complete_rows(n):
    k = 0
    sum = 0
    while sum <= n:
        k += 1
        sum = (k * (k + 1)) // 2
    return k - 1
n=5
result=find_complete_rows(n)
print(result)
```

```python
# ANS 3:
def sortedSquares(nums):
    result = []
    for num in nums:
        result.append(num * num)
    return sorted(result)

nums=[-4,-1,0,3,10]
result=sortedSquares(nums)
print(result)
```

```python
# ANS 4:
def findDisjoint(nums1, nums2):
    set1 = set(nums1)
    set2 = set(nums2)
    result1 = list(set1 - set2)
    result2 = list(set2 - set1)
    return [result1, result2]

nums1 = [1,2,3]
nums2 = [2,4,6]
result=findDisjoint(nums1 ,nums2)
print(result)
```

```python
# ANS 5:
def findDistanceValue(arr1, arr2, d):
    distance = 0
    for num1 in arr1:
        is_valid = True
        for num2 in arr2:
            if abs(num1 - num2) <= d:
                is_valid = False
                break
        if is_valid:
            distance += 1
    return distance

arr1 = [4,5,8]
arr2 = [10,9,1,8]
d = 2
result=findDistanceValue(arr1, arr2, d)
print(result)
```

```python
# ANS 6:
def findDuplicates(nums):
    result = []
    for num in nums:
        num = abs(num)
        if nums[num - 1] > 0:
            nums[num - 1] *= -1
        else:
            result.append(num)
    return result

nums = [4,3,2,7,8,2,3,1]
result=findDuplicates(nums)
print(result)
```

```python
# ANS 7:
def findMin(nums):
    left, right = 0, len(nums) - 1

    while nums[left] > nums[right]:
        mid = (left + right) // 2

        if nums[mid] > nums[right]:
            left = mid + 1
        else:
            right = mid

    return nums[left]

nums = [3,4,5,1,2]
result=findMin(nums)
print(result)
```

```python
# ANS 8:
def findOriginalArray(changed):
    if len(changed) % 2 != 0:
        return []
    
    originalSet = set(changed)
    original = []
    changed.sort()

    for num in changed:
        if num/2 not in originalSet:
            return []
        originalSet.remove(num/2)
        original.append(int(num/2))
    
    return original
```
